```python
import pandas as pd

from sentence_transformers import SentenceTransformer
from bertopic import BERTopic
from bertopic.backend._utils import select_backend
```

```python
imdb_df = pd.read_csv('data/imdb_encoded_with_topics.csv')
```

```python
imdb_not_encoded = pd.read_csv('data/imdb_not_encoded.csv')
```

```python
imdb_clean = pd.merge(imdb_not_encoded[['synopsis', 'link']], imdb_df[['title', 'link', 'topic']], on='link')
```

```python
sentence_model = SentenceTransformer("paraphrase-MiniLM-L6-v2")
model = select_backend(sentence_model)
my_model = BERTopic.load('my_model_without_embedding', embedding_model=model)
```

```python
search_term = 'movie about animals'
topics, probs = my_model.find_topics(search_term)
topic_name = my_model.topic_names[topics[0]]
topic_name = topic_name.replace('_', ' ')
print(f'With {probs[0].round(3)} probability, the search term "{search_term}" falls into "{topic_name}" topic.')
```

```python
topic_df = imdb_clean[imdb_clean.topic == topics[0]][['title', 'synopsis', 'link']]
topic_df
```
